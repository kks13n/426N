```python
%%html
<!DOCTYPE html>
<html lang="zh">
<head>
    <meta charset="UTF-8">
    <meta name="viewport" content="width=device-width, initial-scale=1.0">
    <title>圖片翻轉</title>
    <style>
        body {
            text-align: center;
            font-family: Arial, sans-serif;
            margin-top: 50px;
        }
        .upload-container {
            margin-bottom: 20px;
        }
        .container {
            width: 290px;
            height: 440px;
            margin: auto;
            perspective: 1000px; /* 3D 透視效果 */
        }
        .card {
            width: 100%;
            height: 100%;
            position: relative;
            transform-style: preserve-3d;
            transition: transform 0.6s;
            display: flex;
            justify-content: center;
            align-items: center;
        }
        .flipped {
            transform: rotateY(180deg);
        }
        .side {
            width: 100%;
            height: 100%;
            position: absolute;
            top: 0;
            left: 0;
            backface-visibility: hidden; /* 隱藏背面 */
            object-fit: cover; /* 確保圖片填滿 */
            border-radius: 10px; /* 讓圖片更美觀，可自行移除 */
        }
        .back {
            transform: rotateY(180deg);
        }
    </style>
</head>
<body>

    <div class="upload-container">
        <label>上傳正面圖片：</label>
        <input type="file" id="frontUpload" accept="image/*"><br><br>
        <label>上傳反面圖片：</label>
        <input type="file" id="backUpload" accept="image/*">
    </div>

    <div class="container">
        <div class="card" id="card">
            <img src="" id="frontImage" class="side front" alt="正面">
            <img src="" id="backImage" class="side back" alt="反面">
        </div>
    </div>

    <script>
        let isFlipped = false;

        function flipCard() {
            const card = document.getElementById("card");
            isFlipped = !isFlipped;
            card.classList.toggle("flipped", isFlipped);
        }

        document.getElementById("card").addEventListener("click", flipCard);

        function handleImageUpload(inputElement, imageElementId) {
            const file = inputElement.files[0];
            if (file) {
                const reader = new FileReader();
                reader.onload = function(e) {
                    document.getElementById(imageElementId).src = e.target.result;
                };
                reader.readAsDataURL(file);
            }
        }

        document.getElementById("frontUpload").addEventListener("change", function() {
            handleImageUpload(this, "frontImage");
        });

        document.getElementById("backUpload").addEventListener("change", function() {
            handleImageUpload(this, "backImage");
        });
    </script>

</body>
</html>
```
